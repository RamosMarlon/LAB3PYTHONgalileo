# 4. Dado los Arrays

##### a = [1, 4, 2, 7] y b = [1, 3, 2, 9],
##### obtenga la media aritmética de la diferencia (a-b)

```python
# Definimos los Arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia entre a y b
diferencia = a - b
print("La diferencia es: ",diferencia)

# Calcular la media aritmética de la diferencia
media_diferencia = np.mean(diferencia)
print("La media aritmética de la diferencia es:", media_diferencia)
```

```python
import numpy as np

# Coordenadas iniciales
coordenadas_iniciales = np.array([[1.33, 4.5],
                                  [30.0, 10.7],
                                  [70.2, 0.5]])

# Coordenadas a agregar
coordenadas_nuevas = np.array([[37.1, -3.6]])

# Agregar las coordenadas nuevas al array existente
nuevo_array = np.concatenate((coordenadas_iniciales, coordenadas_nuevas), axis=0)
print('nuevo_array',nuevo_array)
# Mostrar las dimensiones del nuevo array
print(f"Dimensiones del nuevo array: {nuevo_array.shape}")
```

### 2.5 Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
municipios_granada = df[df['Provincia'] == 'Granada']
media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_estandar_poblacion = municipios_granada['Poblacion'].std()
maximo_poblacion = municipios_granada['Poblacion'].max()
minimo_poblacion = municipios_granada['Poblacion'].min()
print("Estadísticas de población de la provincia de Granada:")
print("Media:", media_poblacion)
print("Mediana:", mediana_poblacion)
print("Desviación estándar:", desviacion_estandar_poblacion)
print("Valor máximo:", maximo_poblacion)
print("Valor mínimo:", minimo_poblacion)
```

### 2.6 Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
import matplotlib.pyplot as plt

# Crear un histograma para cada provincia
for provincia in df['Provincia'].unique():
    # Filtrar los municipios de la provincia actual
    municipios_provincia = df[df['Provincia'] == provincia]
   
    # Crear el histograma
    plt.figure(figsize=(10, 6))
    plt.hist(municipios_provincia['Poblacion'], bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma de población de municipios en {provincia}')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

### 2.7 Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
import pandas as pd

# Seleccionar al azar cincuenta municipios sin repetición
municipios_aleatorios = df['Municipio'].sample(n=50, replace=False)

# Filtrar los datos para obtener solo los cincuenta municipios seleccionados
municipios_seleccionados = df[df['Municipio'].isin(municipios_aleatorios)]

# Calcular el promedio y la desviación estándar de la población de estos cincuenta municipios
promedio_poblacion = municipios_seleccionados['Poblacion'].mean()
desviacion_estandar_poblacion = municipios_seleccionados['Poblacion'].std()

# Mostrar los resultados
print("Promedio de la población de los cincuenta municipios seleccionados:", promedio_poblacion)
print("Desviación estándar de la población de los cincuenta municipios seleccionados:", desviacion_estandar_poblacion)
```

### 2.8 Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
import numpy as np

# Coordenadas de Madrid
madrid_x, madrid_y = df[df['Municipio'] == 'Madrid'][['X', 'Y']].values[0]

# Calcular la distancia euclidiana entre Madrid y cada municipio
df['Distancia_Madrid'] = np.sqrt((df['X'] - madrid_x)**2 + (df['Y'] - madrid_y)**2)

# Excluir Madrid y encontrar el municipio más cercano y más lejano
municipio_mas_cercano = df[df['Municipio'] != 'Madrid']['Municipio'].iloc[df['Distancia_Madrid'].idxmin()]
municipio_mas_lejano = df['Municipio'].iloc[df['Distancia_Madrid'].idxmax()]

# Mostrar los resultados
print("Municipio más cercano a Madrid:", municipio_mas_cercano)
print("Municipio más lejano a Madrid:", municipio_mas_lejano)
```
